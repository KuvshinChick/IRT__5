10. Найти какое-либо изображение в сети Интернет. Создать ноутбук, в котором будет отображено выбранное изображение средствами библиотеки matplotlib по URL из сети Интернет.


```python
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image
import requests
from io import BytesIO

# Create a new figure, or activate an existing figure.
plt.figure(figsize=(12, 12))
# get request
url = requests.get('https://sun2-20.userapi.com/impg/7ICnfKbUV3dMJ8h_4k-EXALVb1LTph6OQcG19g/aAkPjhQp1M0.jpg?size=1440x1920&quality=95&sign=476793c39d9921c898a2840fe2c2ce00&type=album')
# response.content returns the content of the response, in bytes
img = Image.open(BytesIO(url.content))
# Display data as an image, i.e., on a 2D regular raster.
plt.imshow(img)
```
