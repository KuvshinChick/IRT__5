##### Построить график зависимости коэффициента диффузии D водорода от температуры T в интервале 100 ≤ T ≤ 600 К через каждые 100 К при p = const = 100 кПа.

##### Дано:
μ = 2·10-3 кг/моль
100 ≤ T ≤ 600 К
ΔТ = 100 К
p = const = 100 кПа = 1·105  Па
σh2 = 0,23*10^-9 м

##### Найти:
D =f(T)  - ?

##### Решение задачи:
Найдем коэффициент диффузии:
$$
\begin{align}
D = \sqrt\frac{1}{3}\times{\bar v}{\bar λ} \\
\end{align}
$$

Для этого выразим среднюю арифметическую скорость и среднюю длину свободного пробеа молекул:
$$
\begin{align}
\bar v = \sqrt{\frac{8\times R\times T}{3\times \pi\times μ}} \\
\end{align}
$$

$$
\begin{align}
\bar λ= \frac{1}{\sqrt{2} \times \pi \times σ^2 \times n}\\
\end{align}
$$

Концентрация молекул:
$$
\begin{align}
n = \frac{p}{ k \times T}\\
\end{align}
$$

Тогда:
$$
\begin{align}
D = \frac{1}{3}\sqrt{\frac{8\times R\times T}{3\times \pi\times μ}} \times \frac{k \times T}{\sqrt{2} \times \pi \times σ^2 \times n}
\end{align}
$$

Подставив значения в формулу, мы получим:
$$
\begin{align}
D = 2,01 \times 10^-8 \times T^\frac{3}{2} \frac{м^2}{с}  \\
\end{align}
$$
##### Построим график функции:

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

# Список значений температуры в Кельвинах
T = np.array([100, 200, 300, 400, 500, 600])
D = T.copy()
# Заполним список значений зависимости коэффициента диффузии D водорода от температуры
D = 2.01*10**(-8) * D**(3/2)
# Построение графика
plt.title("График зависимости коэффициента диффузии D водорода от температуры T") # заголовок
plt.xlabel("T, К") # ось абсцисс
plt.ylabel("D, м^2/с") # ось ординат
plt.grid() # включение отображение сетки
plt.plot(T, D) # построение графика
```
