Цена акций Tesla [2005-2023]

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from itertools import islice

# Открытие файла только для чтения
with open('Tesla_stock_Price.csv', 'r', newline='', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    # Списки
    Data  = []
    High_price = []
    # Пропуск заголовка
    next(reader)
    # Заполнение списков
    for row in islice(reader, 10):
        if row[3] != '':
            Data.append(row[0])
            High_price.append(float(row[3]))

    plt.title('Самая высокая цена акций в приведенные даты:', fontsize=14)
    plt.xlabel("Стоимость")
    plt.ylabel("Дата")
    plt.barh(Data, High_price, color="g")
```
