https://www.kaggle.com/datasets/themrityunjaypathak/tesla-stock-price-2005-2023
Цена акций Tesla [2005-2023]

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from itertools import islice

# Открытие файла только для чтения
with open('Tesla_stock_Price.csv', 'r', newline='', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    # Списки и переменные
    labels = ["> 15 $", " <= 15 $"]
    val= []
    price_1 = 0
    price_2 = 0
    # Пропуск заголовка
    next(reader)
    # Заполнение списков
    for row in islice(reader, 30):
        if float(row[3]) > 15:
            price_1 += 1
        else:
            price_2 += 1
    val.append(price_1)
    val.append(price_2)

    plt.title('Стоимость акций за апрель: \n')
    plt.pie(val, labels=labels, wedgeprops=dict(width=0.5))
```
